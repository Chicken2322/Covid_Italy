```
#Load and compile the model.

import tensorflow as tf 
model = tf.keras.models.load_model('/content/drive/My Drive/Colab Notebooks/Covid_model.h5')
model.compile(loss='binary_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])
```

```
#Importing the necessary files

from keras.preprocessing import image
from keras.models import load_model
from keras.applications.vgg16 import preprocess_input
import matplotlib.pyplot as plt
import numpy as np
```

```
#Testing the model with a new data

img = image.load_img('/content/drive/My Drive/Colab Notebooks/Data/Normal/IM-0255-0001.jpeg', target_size=(224, 224)) 
imgplot = plt.imshow(img)
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
img_data = preprocess_input(x)
classes = model.predict(img_data)
New_pred = np.argmax(classes, axis=1)
if New_pred==[1]:
  print('Prediction: Normal')
else:
  print('Prediction: Corona')
```
